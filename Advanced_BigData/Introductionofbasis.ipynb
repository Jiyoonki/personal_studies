```python
my_counter = dict()

for i in range(100):
    r = random.randint(1,6)
    
    if r in my_counter.keys():
        my_counter[r] = my_counter[r]+1
    else:
        my_counter[r] = 1
    
#     print(r)
```

```python
print(my_counter)
```

# map function
